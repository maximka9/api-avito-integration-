```python
import pandas as pd

# Функции для расчета наценки и округления
def calculate_markup_general(price):
    if price < 4000:
        return round(1.07 * price)
    elif 4000 <= price < 8000:
        return round(1.08 * price)
    elif 8000 <= price < 10000:
        return round(1.10 * price)
    elif 10000 <= price < 12000:
        return round(1.12 * price)
    elif 12000 <= price < 20000:
        return round(1.13 * price)
    elif 20000 <= price < 30000:
        return round(1.15 * price)
    elif price >= 30000:
        return round(1.17 * price)

def calculate_markup_maxxis_prinx(price):
    if price < 8000:
        return round(1.10 * price)
    elif 8000 <= price < 12000:
        return round(1.14 * price)
    elif 12000 <= price < 20000:
        return round(1.15 * price)
    elif 20000 <= price < 30000:
        return round(1.17 * price)
    elif price >= 30000:
        return round(1.20 * price)

def round_price(price):
    return round(price / 50) * 50

# Загрузка данных из файла
file_path = 'tyres_catalog_avito_concatenated.xlsx'
df = pd.read_excel(file_path)

# Применение правил изменения цен для общего случая и округление
df.loc[~df['Brand'].isin(['Maxxis', 'Prinx']), 'Price'] = df.loc[~df['Brand'].isin(['Maxxis', 'Prinx']), 'Price'].apply(calculate_markup_general).astype(int)
# Применение правил изменения цен для брендов Maxxis и Prinx и округление
df.loc[df['Brand'].isin(['Maxxis', 'Prinx']), 'Price'] = df.loc[df['Brand'].isin(['Maxxis', 'Prinx']), 'Price'].apply(calculate_markup_maxxis_prinx).astype(int)

# Округление до ближайшего 50 или 100
df['Price'] = df['Price'].apply(round_price)

# Сохранение обновленной таблицы в Excel
updated_file_path = 'tyres_catalog_avito_updated.xlsx'
df.to_excel(updated_file_path, index=False)
print(f"Файл {updated_file_path} успешно обновлен с учетом новых цен в столбце Price.")
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Загрузка таблиц из Excel файлов
df_new = pd.read_excel('tyres_catalog_avito_updated.xlsx')
df_old = pd.read_excel('tyres_old.xlsx')

# Приведение столбцов Id и Title к строковому типу
df_new['Id'] = df_new['Id'].astype(str)
df_old['Id'] = df_old['Id'].astype(str)
df_new['Title'] = df_new['Title'].astype(str)
df_old['Title'] = df_old['Title'].astype(str)

# 1. Добавить столбцы InternetCalls и CallsDevices в старую таблицу
df_old['InternetCalls'] = 'Да'
df_old['CallsDevices'] = '106575697'

# 2. Удалить столбцы TypeID и TireYear из старой таблицы
df_old.drop(columns=['TypeID', 'TireYear'], errors='ignore', inplace=True)

# 3. В столбце Quantity значение «1» заменить на "за 1 шт." в старой таблице
df_old['Quantity'] = df_old['Quantity'].replace(1, 'за 1 шт.')

# 4. В столбце Brand значение «КАМА» заменить на "КАМА (Нижнекамский шинный завод)" и 'Nokian' на 'Nokian Tyres' в старой таблице
df_old['Brand'] = df_old['Brand'].replace({'КАМА': 'КАМА (Нижнекамский шинный завод)', 'Nokian': 'Nokian Tyres', 'CONTINENTAL Tires': 'Continental', 'Bfgoodrich': 'BFGoodrich'})

# Приведение столбца RimDiameter к строковому типу для обеих таблиц
df_new['RimDiameter'] = df_new['RimDiameter'].astype(str)
df_old['RimDiameter'] = df_old['RimDiameter'].astype(str)

# Проверка наличия Id из df_old в df_new и установка статуса
df_old['AvitoStatus'] = df_old['Id'].apply(lambda x: 'Снято с публикации' if x not in df_new['Id'].values else 'Активно')

# Проверка наличия Title из df_old в df_new и обновление статуса, если товар найден по Title
df_old.loc[df_old['Title'].isin(df_new['Title']), 'AvitoStatus'] = 'Активно'

# Обновление цены, если она отличается между таблицами
df_old = df_old.merge(df_new[['Title', 'Price']], on='Title', how='left', suffixes=('', '_new'))
df_old['Price'] = df_old.apply(lambda row: row['Price_new'] if pd.notna(row['Price_new']) else row['Price'], axis=1)
df_old.drop(columns=['Price_new'], inplace=True)

# Добавление недостающих строк из новой таблицы в старую
df_combined = pd.concat([df_old, df_new[~df_new['Title'].isin(df_old['Title'])]])

# Удаление дубликатов по комбинации столбцов
columns_to_check = ['Brand', 'LoadIndex', 'Model', 'RimDiameter', 'SpeedIndex', 'TireAspectRatio', 'TireSectionWidth']
df_combined.drop_duplicates(subset=columns_to_check, keep='first', inplace=True)

# Удаление дубликатов по Id
df_combined.drop_duplicates(subset=['Id'], keep='first', inplace=True)

# Добавление столбца DateEnd, если его нет
if 'DateEnd' not in df_combined.columns:
    df_combined['DateEnd'] = None

# Получение текущей даты и добавление одного месяца
current_date = datetime.now()
date_plus_one_month = (current_date + timedelta(days=30)).strftime('%Y-%m-%d')

# Обновление столбца DateEnd в df_combined
df_combined['DateEnd'] = df_combined['AvitoStatus'].apply(
    lambda status: '2024-04-11' if status == 'Снято с публикации' else date_plus_one_month
)

# Сохранение объединенной таблицы в новый файл Excel
df_combined.to_excel('tyres_combined.xlsx', index=False)
print("Объединенные данные успешно сохранены в файл tyres_combined.xlsx")
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import requests

# Загрузка таблицы с шинами
tyres_df = pd.read_excel('tyres_combined.xlsx')

# Загрузка XML-файла по ссылке
url = 'https://autoload.avito.ru/format/tyres_make.xml'
response = requests.get(url)
response.raise_for_status()  # Проверка на успешность запроса

# Парсинг XML данных
root = ET.fromstring(response.content)

# Извлечение всех моделей шин из XML
xml_models = set()
for make in root.findall('make'):
    for model in make.findall('model'):
        xml_models.add(model.get('name'))

# Извлечение моделей шин из таблицы
table_models = set(tyres_df['Model'])

# Поиск моделей, которых нет в XML
missing_models = table_models - xml_models

# Вывод отсутствующих моделей
print("Модели, которых нет в XML:")
for model in missing_models:
    print(model)
```
